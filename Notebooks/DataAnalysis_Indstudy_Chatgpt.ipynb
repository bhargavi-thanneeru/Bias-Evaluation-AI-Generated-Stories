```python
# Define the mapping based on keywords
keyword_genre_mapping = {
    'Fairy Tale': ['Fairy'],
    'Fantasy': ['Fantasy'],
    'Children\'s Literature': ['Children', 'Literature'],
    'Adventure': ['Adventure'],
    'Science Fiction': ['Science'],
    'Inspirational': ['Inspirational'],
    'Family & Drama': ['Family', 'Drama'],
    'Educational': ['Educational'],
    'Moral & Inspirational': ['Moral', 'Inspirational'],
    'Friendship & Social': ['Friendship'],
    'Cultural & Diversity': ['Cultural'],
    'Heartwarming & Inspirational': ['Heartwarming'],
    'Action & Adventure': ['Action'],
    'Fiction & Literature': ['Fiction']
}

# Function to categorize genres based on keywords
def categorize_genre(genre_description):
    # Normalize the genre description
    normalized_genre = genre_description.lower()
    for category, keywords in keyword_genre_mapping.items():
        if any(keyword.lower() in normalized_genre for keyword in keywords):
            return category
    return "Other"  # Return 'Other' if no keywords match

# Apply the function to categorize genres
data_clean['Genre'] = data_clean['Genre'].apply(categorize_genre)

# Display the unique genres to check the final categorization
unique_genres = data_clean['Genre'].unique()
print("Categorized Genres in the Dataset:")
print(unique_genres)
```

```python

# Summarization of gender bias explaination
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.cluster.util import cosine_distance
import numpy as np
import networkx as nx

# Function to calculate sentence similarity based on cosine distance
def sentence_similarity(sent1, sent2, stopwords=None):
    if stopwords is None:
        stopwords = []

    sent1 = [w.lower() for w in sent1]
    sent2 = [w.lower() for w in sent2]

    all_words = list(set(sent1 + sent2))

    vector1 = [0] * len(all_words)
    vector2 = [0] * len(all_words)

    # build the vector for the first sentence
    for w in sent1:
        if w in stopwords:
            continue
        vector1[all_words.index(w)] += 1

    # build the vector for the second sentence
    for w in sent2:
        if w in stopwords:
            continue
        vector2[all_words.index(w)] += 1

    return 1 - cosine_distance(vector1, vector2)

# Function to generate summary using TextRank algorithm
def generate_summary(sentences, top_n=5):
    stop_words = stopwords.words('english')
    summarize_text = []

    # Tokenize sentences into words
    sentence_tokens = [sent_tokenize(sentence) for sentence in sentences]
    sentence_tokens = [y for x in sentence_tokens for y in x]

    # Vectorize the sentences
    sentence_vectors = []
    for sentence in sentence_tokens:
        sentence_tokens = nltk.word_tokenize(sentence)
        sentence_tokens = [token.lower() for token in sentence_tokens]
        sentence_vectors.append(sentence_tokens)

    # Calculate sentence scores based on similarity
    similarity_matrix = np.zeros([len(sentence_tokens), len(sentence_tokens)])
    for i in range(len(sentence_tokens)):
        for j in range(len(sentence_tokens)):
            if i == j:
                continue
            similarity_matrix[i][j] = sentence_similarity(sentence_vectors[i], sentence_vectors[j], stop_words)

    # Apply PageRank algorithm
    scores = nx.pagerank(nx.from_numpy_array(similarity_matrix))

    # Sort the sentence scores and select top sentences
    ranked_sentences = sorted(((scores[i], s) for i, s in enumerate(sentence_tokens)), reverse=True)
    for i in range(top_n):
        summarize_text.append("".join(ranked_sentences[i][1]))

    return ". ".join(summarize_text)

# Filter the DataFrame for records with gender bias boolean as true
gender_bias_true_df = data_clean[data_clean['Gender Bias Boolean'] == True]

# Extract the gender bias explanations from the filtered DataFrame
gender_bias_explanations = gender_bias_true_df['Gender Bias'].tolist()
print(len(gender_bias_explanations))
print(gender_bias_explanations)

# Generate summary for the gender bias explanations
summary = generate_summary(gender_bias_explanations)

# Print or store the summarized explanation
print("Summarized Explanation for Gender Bias:")
print(summary)

```
